```python
submit_time = pd.to_datetime(df['submit_time'], errors='coerce')
df = df[submit_time.dt.year == 2023] # only keep 2023 data
df_=df.copy()
```

# Drop NANS

```python
# Total number of rows in the original DataFrame
total_rows = df_.shape[0]

# Drop rows with NaNs
df_completed_without_nans = df_.dropna()

# Number of rows after dropping NaNs
rows_without_nans = df_completed_without_nans.shape[0]

# Number of rows dropped
rows_dropped = total_rows - rows_without_nans

# Percentage of rows dropped
percentage_dropped = (rows_dropped / total_rows) * 100

print(f"Total number of rows: {total_rows}")
print(f"Number of rows after dropping NaNs: {rows_without_nans}")
print(f"Number of rows dropped: {rows_dropped}")
print(f"Percentage of rows dropped: {percentage_dropped:.2f}%")
```

## Delete rows having runtime <= 0 

```python
df_completed_without_nans_zero = df_completed_without_nans[df_completed_without_nans['run_time'] > 0]
print("Shape before:", df_completed_without_nans.shape)
print("Shape after removing zero run_time:", df_completed_without_nans_zero.shape)
```

## Delete unuseful columns

```python
columns_to_drop = ['state','start_time']
df_completed_without_nans_zero_drop = df_completed_without_nans_zero.drop(columns=columns_to_drop)

print("Columns dropped successfully.")
print("Remaining columns:", df_completed_without_nans_zero_drop.columns)
```

## Delete Duplicates

```python
df_completed_without_nans_zero_drop_duplicates = (
    df_completed_without_nans_zero_drop
    .drop_duplicates()  # remove full duplicates
    .reset_index(drop=True)  # reset index after
)
```

```python
print(len(df_completed_without_nans_zero_drop) - len(df_completed_without_nans_zero_drop_duplicates))
```

```python
import os

# Define the directory name
directory_name = "datasets_after_clean_1"

# Check if the directory exists, if not, create it
if not os.path.exists(directory_name):
    os.makedirs(directory_name)
    print(f"Directory '{directory_name}' created.")
else:
    print(f"Directory '{directory_name}' already exists.")
```

```python
def classer_job(runtime):
    if runtime <= 600:# 10 minutes
        return 'short'
    elif runtime <= 2400: # 30 minutes
        return 'medium'
    else:
        return 'long'

df_completed_without_nans_zero_drop_duplicates['job_length'] = df_completed_without_nans_zero_drop_duplicates['run_time'].apply(classer_job)
```

```python
df_completed_without_nans_zero_drop_duplicates.to_csv('datasets_after_clean_1/eagle_data_all_completed.csv', index=False)
```
